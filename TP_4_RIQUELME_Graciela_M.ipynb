```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV
file_path = '/content/drive/MyDrive/Colab Notebooks/Crimenes/Crimenes.csv'
data = pd.read_csv(file_path)

# Imprime los nombres de las columnas para verificar
print(data.columns)

# Asegúrate de que los nombres de las columnas coincidan con tus datos reales.
# Ajusta los nombres de las columnas según sea necesario.

estados = data['Estado']
asaltos = data['Asesinatos']
violaciones = data['Asaltos']
asesinatos = data['Violaciones']

# Ajusta el nombre de la columna según los datos reales.
# Usa el nombre exacto que se muestra al imprimir las columnas.

poblacion_urbana = data['Poblacion Urbana']

# Crear una figura con un tamaño específico

plt.figure(figsize=(25, 5))

# Crear subgráficos para cada conjunto de datos
plt.subplot(1, 4, 1)
plt.bar(estados, asaltos)
plt.title('Asaltos')
plt.xticks(rotation=90, fontsize=8)

plt.subplot(1, 4, 2)
plt.bar(estados, violaciones)
plt.title('Violaciones')
plt.xticks(rotation=90, fontsize=8)

plt.subplot(1, 4, 3)
plt.bar(estados, asesinatos)
plt.title('Asesinatos')
plt.xticks(rotation=90, fontsize=8)

plt.subplot(1, 4, 4)
plt.bar(estados, poblacion_urbana)
plt.title('Población Urbana')
plt.xticks(rotation=90, fontsize=8)

# Ajustar el diseño para evitar solapamientos
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# GRAFICA de ASALTOS:
# Calcular el promedio de asaltos
promedio_asaltos = asaltos.mean()

# Verificar si algún estado tiene un número de asaltos significativamente superior al promedio
estados_con_mas_asaltos = estados[asaltos > promedio_asaltos]

# Visualizar el gráfico
plt.figure(figsize=(20, 5))
plt.bar(estados, asaltos)
plt.xticks(rotation=90)
plt.title('Asaltos')

# Interpretación
print(f"Promedio de asaltos: {promedio_asaltos}")
print(f"Estados con más asaltos que el promedio: {estados_con_mas_asaltos.tolist()}")
```

```python
# GRAFICO de VIOLACIONES:
# Calcular el promedio de violaciones
promedio_violaciones = violaciones.mean()

# Identificar estados con un número de violaciones significativamente superior al promedio
estados_con_mas_violaciones = estados[violaciones > promedio_violaciones]

# Visualizar el gráfico
plt.figure(figsize=(20, 5))
plt.bar(estados, violaciones)
plt.title('Violaciones')

plt.xticks(rotation=90)


# Interpretación
print(f"Promedio de violaciones: {promedio_violaciones}")
print(f"Estados con más violaciones que el promedio: {estados_con_mas_violaciones.tolist()}")
```

```python
# GRAFICO de ASESINATOS:
# Supongamos que ya has cargado el archivo y tienes acceso a los datos.
# Calcular el promedio de asesinatos
promedio_asesinatos = asesinatos.mean()

# Identificar estados con un número de asesinatos significativamente superior al promedio
estados_con_mas_asesinatos = estados[asesinatos > promedio_asesinatos]
# Visualizar el gráfico
plt.figure(figsize=(20, 5))
plt.xticks(rotation=90)
plt.bar(estados, asesinatos)
plt.title('Asesinatos')

# Interpretación
print(f"Promedio de asesinatos: {promedio_asesinatos}")
print(f"Estados con más asesinatos que el promedio: {estados_con_mas_asesinatos.tolist()}")
```

```python
# Este bloque de código escala los datos utilizando StandardScaler,
# y luego crea el dendrograma de tasa de criminalidad utilizando el método de vinculación de "fuga única".
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import StandardScaler
# Escalar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(crimenes[['Asesinatos', 'Asaltos', 'Poblacion Urbana', 'Violaciones']])
# Crear el dendrograma de tasa de criminalidad con el método de vinculación de "fuga única"
plt.figure(figsize=(20, 5))
plt.title("Dendrogramas de tasa de criminalidad")
dend = sch.dendrogram(sch.linkage(scaled_data, method='single'))
plt.xlabel('Tasa de criminalidad')
plt.ylabel('Distancias euclidianas')
plt.show()
```

```python
## Este bloque crea el dendrograma de tasa de criminalidad utilizando
# el método de vinculación completa en los datos escalados.
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
# Crear el dendrograma de tasa de criminalidad con el método de vinculación completa
plt.figure(figsize=(20, 5))
plt.title("Dendrogramas de tasa de criminalidad")
dend = sch.dendrogram(sch.linkage(scaled_data, method='complete'))
plt.xlabel('Tasa de criminalidad')
plt.ylabel('Distancias euclidianas')
plt.show()
```

```python
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
# Crear el dendrograma de tasa de criminalidad con el método de vinculación promedio
plt.figure(figsize=(20, 5))
plt.title("Dendrogramas de tasa de criminalidad")
dend = sch.dendrogram(sch.linkage(scaled_data, method='average'))
plt.xlabel('Tasa de criminalidad')
plt.ylabel('Distancias euclidianas')
plt.show()
```

### En este código:Primero, importamos la biblioteca necesaria AgglomerativeClustering de scikit-learn. Ajustamos el modelo de agrupación aglomerativa con el número óptimo de grupos, que en este caso es 3 como se determinó a partir del dendrograma. Ajustamos y predecimos las etiquetas del grupo. Asignamos las etiquetas del grupo al DataFrame original "crimenes". Mostramos qué estado pertenece a qué grupo en el DataFrame resultante.

```python
# podemos calcular el número óptimo de grupos utilizando el método de vinculación promedio
# y el algoritmo Agglomerative Clustering:
```

```python
# Importar la biblioteca necesaria
from sklearn.cluster import AgglomerativeClustering
# Ajustar la agrupación aglomerativa con el número óptimo de grupos
n_clusters = 3  # Determinado a partir del dendrograma
AC = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='ward')
# Ajustar y predecir las etiquetas del grupo
y_pred = AC.fit_predict(crimenes[['Asesinatos', 'Asaltos', 'Poblacion Urbana', 'Violaciones']])
# Asignar las etiquetas del grupo al DataFrame original
crimenes['Cluster Labels'] = y_pred
# Mostrar qué estado pertenece a qué grupo
print(crimenes[['Estado', 'Cluster Labels']])
```

```python
# El siguiente código utiliza gráficos de caja (boxplots) para mostrar las tasas de asesinato,
# violaciónes y asalto en diferentes grupos. Luego,
# se muestran las observaciones y se obtienen los estados que pertenecen a cada zona
# (segura, peligrosa y moderada).
```

```python
import seaborn as sns
# Mostrar tasas de asesinato por etiqueta de grupo
plt.figure(figsize=(10, 5))
sns.boxplot(x='Cluster Labels', y='Asesinatos', data=crimenes)
plt.title('Tasa de Asesinato por Etiqueta de Grupo')
plt.xlabel('Etiqueta de Grupo')
plt.ylabel('Tasa de Asesinato')
plt.show()
# Mostrar tasas de violación por etiqueta de grupo
plt.figure(figsize=(10, 5))
sns.boxplot(x='Cluster Labels', y='Violaciones', data=crimenes)
plt.title('Tasa de Violaciones por Etiqueta de Grupo')
plt.xlabel('Etiqueta de Grupo')
plt.ylabel('Tasa de Violación')
plt.show()
# Mostrar tasas de asalto por etiqueta de grupo
plt.figure(figsize=(10, 5))
sns.boxplot(x='Cluster Labels', y='Asaltos', data=crimenes)
plt.title('Tasa de Asaltos por Etiqueta de Grupo')
plt.xlabel('Etiqueta de Grupo')
plt.ylabel('Tasa de Asaltos')
plt.show()
# Observaciones
print("Observaciones:")
print("• Las ciudades del Clúster-0 parecen ser zonas seguras donde hay relativamente menos asesinatos, agresiones y violaciones.")
print("• Las ciudades del Grupo 1 parecen tener tasas de criminalidad más altas y pueden considerarse como Zona de Peligro.")
print("• Las ciudades del Grupo 3 parecen tener tasas de criminalidad moderadas en comparación con otras zonas y pueden denominarse Zona Moderada.")
# Obtener estados de cada zona
Safe_Zone = crimenes[crimenes['Cluster Labels'] == 0]['Estado'].unique()
Danger_Zone = crimenes[crimenes['Cluster Labels'] == 1]['Estado'].unique()
Moderate_Zone = crimenes[crimenes['Cluster Labels'] == 2]['Estado'].unique()
print("Zona segura:")
print(Safe_Zone)
print("Zona peligrosa:")
print(Danger_Zone)
print("Zona moderada:")
print(Moderate_Zone)
```

```python
# Importar la biblioteca necesaria
from sklearn.cluster import AgglomerativeClustering

# Ajustar la agrupación aglomerativa con el número óptimo de grupos
n_clusters = 3  # Determinado a partir del dendrograma
AC = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='ward')

# Ajustar y predecir las etiquetas del grupo
y_pred = AC.fit_predict(crimenes[['Asesinatos', 'Asaltos', 'Poblacion Urbana', 'Violaciones']])

# Asignar las etiquetas del grupo al DataFrame original
crimenes['Cluster Labels'] = y_pred

# Mapear las etiquetas del clúster a las categorías deseadas
crimenes['Zona de Peligrosidad'] = crimenes['Cluster Labels'].map({0: 'Baja', 1: 'Alta', 2: 'Mediana'})

# Mostrar qué estado pertenece a qué grupo y su zona de peligrosidad
print(crimenes[['Estado', 'Cluster Labels', 'Zona de Peligrosidad']])
```
